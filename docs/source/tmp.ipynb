```python
from PIL import Image, ImageFilter
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = Image.open("logo_wide.png")
x = np.array(image)
x = x[::3, ::3]

c = np.array([255, 255, 255, 0], dtype=np.uint8)
c = np.tile(c, (x.shape[0], 200, 1))

x = np.concatenate((c, x, c), axis=1)

plt.imshow(x)
plt.show()


# save
image = Image.fromarray(x)
image.save("logo_wider.png")
```
